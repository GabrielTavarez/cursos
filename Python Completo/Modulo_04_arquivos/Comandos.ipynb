# Arquivos de biblioteca

Módulos pathlib

* Path.cwd() -> current working directory: retorna o caminho do diretório
* Path(Caminho_em_linux) -> Usar barra normal (/) e não contrabarra (\) e retorna o caminho de qualquer Sist Ope. Caminho pode ser absoluto ou relativo
* caminho.iterdir() -> retorna uma lista de arquivos 
* / -> operado de concatenação de 2 objetos Path
* caminho.exists() -> retorna booleano se algum arquivo já existe ou não no computador

* caminho.mkdir() -> Cria uma nova pasta no caminho se não existir

módulos shutil

* shutil.copy2(arquivo_copy, arquivo_paste) -> arquivos são passados como Paths
* shutil.move(caminho_origem, caminho_mover) -> caminhos passados como path
