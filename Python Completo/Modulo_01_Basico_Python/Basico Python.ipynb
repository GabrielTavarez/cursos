* **FOR**

Variável percorre todos os iten de uma lista. Pode ser um index (range) ou itens diretos de uma lista

```python
for variavel in lista :
    #codigo
for i in range(n) :
    print(i)
for i in [1,2,3,4]:
    print(i)
for i in ["primerio", "segundo", "terceiro"]
    print(i)
```

Pode interromper o loop com brak e continue

```python
for i in range(10) :
    if i > 5:
        break #interrompe imediatamente o for
    print(i)
#1,2,3,4,5
    
for i in range(10) :
    if i < 5:
        continue # para de ler tudo abaixo e volta para o for
    print(i)
#5,6,7,8,9
```

Função enumerate retorna o valor do interator e o valor do indice

```python
for i, item in enumerate(lista)
    print(i) #indice
    print(item) #item lista[i]
```
