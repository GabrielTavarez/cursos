### Passo 4: Importar o arquivo pro nosso computador com o pandas

```python
import pandas as pd

file_path = r'C:\Users\gabri\OneDrive\Documents\Cursos\Python Completo\Mentoria Automação cartão crédito\BankChurners.csv'

clientes_df = pd.read_csv(file_path)

display(clientes_df)
```

### Passo 5: Calcular os indicadores para enviar para a diretoria:

    1 - Quantos clientes Ativos e Quantos Clientes em Disputa Temos?
    2 - Dos Clientes Ativos, como está a distribuição por tipo de cartão
    3 - Qual o tempo médio de permanência e o limite de crédito médio dos nosso clientes (totais e ex-clientes)?

```python
resumo_status = clientes_df.groupby('Attrition_Flag')['Attrition_Flag'].count()
print(resumo_status.to_string())
```

```python
clientes_ativos = clientes_df.loc[clientes_df['Attrition_Flag'] == 'Existing Customer', ['Attrition_Flag', 'Card_Category']]
resumo_cartoes =  clientes_ativos.groupby('Card_Category')['Card_Category'].count()
resumo_cartoes.index.names = ['Categoria do Card - Clientes ativos']


print(resumo_cartoes.to_string())
```

```python
tempo_medio = clientes_df['Months_on_book'].mean()
print(f'Tempo médio: {tempo_medio:.1f} meses')
clientes_inativos = clientes_df.loc[clientes_df['Attrition_Flag'] == 'Attrited Customer', :]
limite_exclientes = clientes_inativos['Credit_Limit'].mean()
print(f'Crédito médio de ex clientes: R${limite_exclientes:.2f}')
```
