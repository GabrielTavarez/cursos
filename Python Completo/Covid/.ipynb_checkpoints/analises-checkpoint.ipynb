```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import statistics
```

```python
#importação de dados

data_dir = 'HIST_PAINEL_COVIDBR_27jan2022'

files = os.listdir(data_dir)

dic_covid_files = {}

#covid = pd.read_csv(data_dir + '\\' + files[0], sep = ';')

for i in range(0, len(files)-1):
    dic_covid_files[files[i]] = pd.read_csv(data_dir + '\\' + files[i], sep = ';')
    dic_covid_files[files[i]]['data'] = pd.to_numeric(dic_covid_files[files[i]]['data'].str.replace('-',''))
    
    #display(dic_covid_files[files[i]])
        
    #covid = pd.concat([covid, pd.read_csv(data_dir + '\\' + files[i], sep = ';')])
    #covid['data'] = pd.to_datetime(covid['data'], format='%Y-%m-%d')
#display(covid[1033892:1033897])
```

```python
covid = pd.concat(list(dic_covid_files.values()))

covid.dropna()

display(covid)
```

```python
print('Max parte 1 :', dic_covid_files[files[0]]['data'].max())
print('Min parte 2 :', dic_covid_files[files[1]]['data'].min())
print('Max parte 2 :', dic_covid_files[files[1]]['data'].max())
print('Min parte 3 :', dic_covid_files[files[2]]['data'].min())
print('Max parte 3 :', dic_covid_files[files[2]]['data'].max())
print('Min parte 4 :', dic_covid_files[files[3]]['data'].min())
print('Max parte 4 :', dic_covid_files[files[3]]['data'].max())
```

```python
covid_brasil = covid[['data', 'casosAcumulado', 'casosNovos', 'obitosAcumulado', 'obitosNovos']].\
                                                                        groupby('data', as_index= False).sum()
```

```python
#calc de média móvel
dias_media = 7

covid_brasil['mediaMovelObitos'] = 0

for i in range(dias_media-1, len(covid_brasil) -1):
    covid_brasil.loc[i, 'mediaMovelObitos'] = statistics.mean(covid_brasil.loc[i-dias_media:i, 'obitosNovos'])

plt.figure(figsize = (15,5))
plt.plot(covid_brasil['data'], covid_brasil['mediaMovelObitos'] )
plt.show()

plt.figure(figsize = (15,5))
plt.plot(covid_brasil['data'], covid_brasil['obitosAcumulado'] )
plt.show()
```
