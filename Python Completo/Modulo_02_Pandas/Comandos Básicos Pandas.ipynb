## Filtragem de valores

Para filtrar valores, é necessário fazer uma condição de acesso de valores:

dataframe[condição_linha, condição_coluna]

na verdade o parâmetro passado é uma lista de Booleanos definindo se a condição é verdadeira ou falsa.

```python
maior2 = vendas_df["Quantidade"] > 2 #lista de True/False
display(vendas_df[maior2][:5])

#pode ser feito direto
#display(vendas_df[ vendas_df["Quantidade"] ])
```

# Método .loc[  ]
Método loc[linha,coluna]
linha e coluna podem ser condições da linha ou coluna, numéros ou intervalos

É interessante para poder alterar valores em um Dataframe

# Pegar 1 linha

```python
linha = vendas_df.loc[1] #linha é uma series
display(linha)
linhas = vendas_df.loc[0:5] #linhas é um dataframe
display(linhas)
```

# Pegar linhas por condição

```python
#apenas linhas da loja norte shopping

linhas = vendas_df.loc[vendas_df['ID Loja']=='Norte Shopping'] #condição dentro dos parênteses
display(linhas[:5])
```

# Pegar Linhas e Colunas específicas

```python
#método loc[linhas,colunas]

vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja']=='Norte Shopping', ['ID Loja','Produto','Quantidade']]
display(vendas_norteshopping_df[:5])
```

# Alterar linhas e colunas por condições

DataFrame.loc[ condição_linha, condição_coluna ] = valor

```python
vendas_df.loc[ vendas_df["ID Loja"]== "Norte Shopping" , "ID Loja"] = "Shopping Norte" 
display(vendas_df.head())
```

# Adicionar uma coluna

```python
vendas_df['Comissão'] = vendas_df['Valor Final']*0.05 #Cria uma coluna Comissão se não existir nenhuma e preenche com o valor recebido

vendas_df.loc[:,'Imposto'] = 0 #adiciona o valor 0 em todas as linhas (:) na coluan de Imposto que foi criada se não existir
display(vendas_df[:5])
```

# Adicionar linhas

```python
vendas_dezembro_df = pd.read_excel("Vendas - Dez.xlsx")
display(vendas_dezembro_df[-5:-1])

vendas_df = vendas_df.append(vendas_dezembro_df) #acrescenta o dataframe de dezembro no dataframe original
display(vendas_df[-5:-1])
```

# Excluir colunas

```python
vendas_df = vendas_df.drop("Imposto", axis = 1) #axis=1 é a coluna, axis = 0 é a linha
vendas_df = vendas_df.drop(0,axis=0) #exclui a coluna 0 do dataframe
display(vendas_df[:5])
```

# Deletar linhas e colunas completamente vazias

```python
vendas_df = vendas_df.dropna(how='all', axis =0) #exclui linhas completamente vazias
vendas_df = vendas_df.dropna(how='all', axis =1) #exclui colunas completamente vazias 
```

# Deletar linhas com valores vazio

```python
vendas_df = vendas_df.dropna() #remove linhas que tem qualquer valor vazio
```

# Preencher valores vazios

```python
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(0) #preenche todos os valores vazios da coluna com o número 0
#vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean()) #preenche todos os valores vazios da coluna com a media
vendas_df['Comissão'] = vendas_df.ffill() #preenche os valores vazios com o último valor real na coluna antes dos NaN

display(vendas_df[:5])
```

# Calcular valores por coluna

```python
transacoes_loja = vendas_df['ID Loja'].value_counts() #conta a quantidade repetições de cada valor na coluna
display(transacoes_loja[:5])
```

# Agrupamentos

```python
#agrupa os elementos da coluna de produtos e soma as outras colunas numéricas
faturamento_produto = vendas_df.groupby('Produto').sum() 

#faturamento_produto = vendas_df[['Produto','Valor Final]]'.groupby('Produto').sum() #seleciona apenas coluna de produto e valor
display(faturamento_produto[:5])
```

## Ordenard DataFrames

```python
display(faturamento_produto.sort_values("Quantidade")[:5]) # ordena em ordem decrescente a partir de Quantidades
```
