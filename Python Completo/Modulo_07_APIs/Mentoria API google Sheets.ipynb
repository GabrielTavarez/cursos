```python
from __future__ import print_function

import os.path

from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError

# If modifying these scopes, delete the file token.json.
SCOPES = ['https://www.googleapis.com/auth/spreadsheets']#.readonly -> apenas leitura de planilhas

# The ID and range of a sample spreadsheet.



def main():
    ##CONEXÃO GOOGLE==============================
    """Shows basic usage of the Sheets API.
    Prints values from a sample spreadsheet.
    """
    creds = None
    # The file token.json stores the user's access and refresh tokens, and is
    # created automatically when the authorization flow completes for the first
    # time.
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json', SCOPES)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES)
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run
        with open('token.json', 'w') as token:
            token.write(creds.to_json())

    try:
        service = build('sheets', 'v4', credentials=creds)

        # Call the Sheets API==================================================
        sheet = service.spreadsheets()
        
        
        #PEGAR VALORES DAS CELULAS
        SAMPLE_SPREADSHEET_ID = '11Xe-GbplX-C74-9V0GChCqjEKefAOC-dYk_qHyKuOcw' #é o id da planilha -> achar no link da planilha
        SAMPLE_RANGE_NAME = 'page1!A1:E' #intervalo de células que quer editar/selecionar
        
        #seleciona intervalo de células
        result = sheet.values().get(spreadsheetId=SAMPLE_SPREADSHEET_ID,
                                    range=SAMPLE_RANGE_NAME).execute()
        
        values = result.get('values', []) #Retorna uma lista onde cada item é uma lista com os itens da linh

        print('A, B:')
        #printa os valores de cada linha
        for row in values:
            # Print columns A and E, which correspond to indices 0 and 4.
            print('%s, %s' % (row[0], row[1]))
        
        #ESCREVE VALORES NA PLANILHA
        values = [
            [
                'Texto escrito por python'
            ],
            # Additional rows ...
        ]
        
        body = {
            'values': values
        }
        
        spreadsheet_id = '11Xe-GbplX-C74-9V0GChCqjEKefAOC-dYk_qHyKuOcw'
        range_name = 'page2!A1:A5'
        value_input_option = 'RAW' #Não escreve fórmulas, apenas textos
        
        
        result = service.spreadsheets().values().update( #método update altera os valores selecionados
            spreadsheetId=spreadsheet_id, range=range_name,
            valueInputOption=value_input_option, body=body).execute()
        
        result = service.spreadsheets().values().append( #método append adiciona valores no fim da lista no intervalo
            spreadsheetId=spreadsheet_id, range=range_name,
            valueInputOption=value_input_option, body=body).execute()

    
    
    
    
    except HttpError as err:
        print(err)
        
```

```python
if __name__ == '__main__':
    main()
```
